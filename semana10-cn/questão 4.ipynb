```python
import numpy as np
```

```python
## definir quantas funções você quer pro polinomio


def gk(x,k):
    if x == k:
        return 1
    else:
        return ((np.sin(np.pi*(x - k)))/(x-k))





dicionarioFuncoes = { 0:gk}
```

```python
## transcrever aqui as tabelas de valores de x e y dados no exercicio

arr_x = np.arange(0,47.01)

def filly(arr_x):
    arr_y = []
    for x in arr_x:
        arr_y.append(np.sin(x))
    return arr_y

arr_y = filly(arr_x)
print(arr_x)
print(arr_y)
```

```python


def fillmg(dicionarioFuncoes, arr_x):
    matrizGtrans = []
    for i in range(12):
        f = dicionarioFuncoes[0]
        tmpArr = []
        for x in arr_x:
            tmpArr.append(f(x,i))
        matrizGtrans.append(tmpArr)
    matrizGtransnp = np.array(matrizGtrans)
    return matrizGtransnp


matrizG = fillmg(dicionarioFuncoes,arr_x)
```
